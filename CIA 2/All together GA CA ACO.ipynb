# Normal

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

```python
def ReLU(x):
    return x * (x > 0)
```

```python
class organism:
    def __init__(self, input_size, hidden_size, output_size,weight1,weight2):
        self.ip = input_size
        self.hp = hidden_size
        self.op = output_size
        self.weight1 = weight1
        self.b1 = np.zeros((1, self.hp))
        self.weight2 = weight2
        self.b2 = np.zeros((1, self.op))

    def forward(self, x):
        self.z1 = np.dot(x, self.weight1) + self.b1
        self.a1 = sigmoid(self.z1)
        self.z2 = np.dot(self.a1, self.weight2) + self.b2
        self.a2 = sigmoid(self.z2)
        return self.a2
```

```python
def loss(y_hat, y):
    return -np.mean(y * np.log(y_hat) + (1 - y) * np.log(1 - y_hat))
```

```python
def train(X, y, nn, epochs, learning_rate):
    for i in range(epochs):
        y_hat = nn.forward(X)
        cost = loss(y_hat, y)
        dz2 = y_hat - y
        dw2 = np.dot(nn.a1.T, dz2)
        db2 = np.sum(dz2, axis=0, keepdims=True)
        dz1 = np.dot(dz2, nn.weight2.T) * (nn.a1 * (1 - nn.a1))
        dw1 = np.dot(X.T, dz1)
        db1 = np.sum(dz1, axis=0)
        nn.weight1 -= learning_rate * dw1
        nn.b1 -= learning_rate * db1
        nn.weight2 -= learning_rate * dw2
        nn.b2 -= learning_rate * db2
        if i % 100 == 0:
            print("Epoch:", i, "Cost:", cost)
```

# Cultural Algo

```python
total_ppln = 10
best = 2
mutation_rate = 0.1
belief = []
ppln_cul = []
```

```python
def initial_deploy_cul(n):
    animals = []
    
    for _ in range(n):
        animals.append(organism(10, 20, 1,np.random.randn(10, 20),np.random.randn(20, 1)))
        
    return animals
    
```

```python
def crossover(parent1, parent2):
    if np.random.rand() < 0.5:
        weight1 = parent1.weight1
        weight2 = parent2.weight2
    else:
        weight1 = parent2.weight1
        weight2 = parent1.weight2
        
    return weight1,weight2
```

```python
def mutate(solution):
    e = np.random.choice(20, np.random.randint(20), replace=False)
    if np.random.rand() < mutation_rate:
        for i in e:
            weight2[i] = weight2[i]+np.random.randn()
    return solution
```

```python
def select(fitness_values,population):
    sorted_indices = np.argsort(fitness_values)[::-1]
    return [population[i] for i in sorted_indices[:best]]
```

```python
def new_belief(soln):
    for i in soln:
        belief.append(i)
    return sorted(belief, key=lambda x: fitness(nn), reverse=True)[:total_ppln]
```

```python
ppln_cul = initial_deploy_cul(total_organisms)
for i in range(100):
    population = []
    for j in range(len(ppln_cul)):
        print("-",end=" ")
        if np.random.rand() < 0.5 and len(belief) > 0:
            parent1, parent2 = np.random.choice(belief, size=2, replace=False)
            weight1,weight2 = crossover(parent1, parent2)
        else:
            weight1 = weight1 + np.random.randn()
            weight2 = weight2 + np.random.randn()
            
        weight2 = mutate(weight2)
        population.append(organism(10, 20, 1,weight1,weight2))

    fitness_values = [fitness(nn) for nn in population]
    selected_solutions = select(fitness_values,population)
    belief = new_belief(selected_solutions)

    cur_soln = belief[0]
    weight1 = cur_soln.weight1
    weight2 = cur_soln.weight2

    if i % 10 == 0:
        print(f"> Generation: {i}, Fitness: {fitness(organism(10,20,1,weight1,weight2))}")
```

# ACO

```python
def create(size,stopper):
    graph = np.full((size,size),np.inf)
    for i in range(size):
        for j in range(size):
            randy = np.random.randint(0,10)
            if randy>stopper and i!=j:
                n = np.random.randint(1,30)
                graph[i][j] = n
                graph[j][i] = n
                
    return graph
```

```python
def phc(graph):
    return np.ones((len(graph),len(graph)),dtype=np.float64)
```

```python
def nextway(graph,ph,curr):
    graph = 1/graph
    bottom = np.dot(graph[curr],ph[curr])
    top = graph[curr]*ph[curr]
    prob = top/bottom

    way = np.cumsum(prob)
    rb = np.random.random()
    
    nextvertex = np.where(way > rb)[0][0]
    
    return nextvertex
```

```python
def traverse(graph,ph,start,end):
    pre = start
    prev = start
    path=[start]
    cost=0
    while(pre!=end):
        pa = nextway(graph,ph,pre)
        while(prev==pa):
            pa = nextway(graph,ph,pre)
        path.append(pa)
        cost+= graph[pa][pre]
        prev=pre
        pre = pa
        
    return path,cost
```
